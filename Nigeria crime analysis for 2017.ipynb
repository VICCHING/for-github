```python
import matplotlib.pyplot as plt
import pandas as pd
file1=r"C:\Users\HP\Downloads\CRIME STATISTICS 2017.xlsx"
df=pd.read_excel(file1, "person")
df
```

```python
df.set_index("STATE", inplace=True)
df
```

```python
df[df["MURDER"]== df["MURDER"].max()]
```

```python
df.groupby("STATE")["MURDER"].max().plot(kind = "barh", figsize=(10,10))
plt.grid(which="both")
plt.title("MURDER CASES IN STATE",fontweight="bold")
plt.xlabel("NUMBER OF MURDER")
plt.ylabel("STATE")
```

```python
df.groupby("STATE")["SUICIDE"].max().plot(kind = "barh", figsize=(10,10))
plt.grid(which="both")
plt.title("SUICIDE CASES IN STATE",fontweight="bold")
plt.xlabel("NUMBER OF SUICIDE")
plt.ylabel("STATE")
```

```python
df.groupby("STATE")["KIDNAPPING"].max().plot(kind = "barh", figsize=(10,8))
plt.grid(which="both")
plt.title("KIDNAPPING CASES IN STATE",fontweight="bold")
plt.xlabel("NUMBER OF SUICIDE")
plt.ylabel("STATE")
```

```python
df.groupby("STATE")["RAPE & INDICENT ASSAULT"].max().plot(kind = "barh", figsize=(10,10))
plt.minorticks_on()
plt.grid(which="both")
plt.title("RAPE AND INDICENT ASSAULT IN STATE",fontweight="bold")
plt.xlabel("RAPE & INDICENT ASSAULT")
plt.ylabel("STATE")
```

```python
file1=r"C:\Users\HP\Downloads\CRIME STATISTICS 2017.xlsx"
Summary=pd.read_excel(file1, "summary2")
Summary
```

```python
Summary.set_index("STATE", inplace=True)
Summary
```

```python
Summary.groupby("STATE")["OFFENCES AGAINST PERSONS"].max().plot(kind = "barh", figsize=(10,10))
plt.grid(which="both")
plt.title("OFFENCES AGAINST PERSONS IN STATE",fontweight="bold")
plt.xlabel("OFFENCES AGAINST PERSONS")
plt.ylabel("STATE")
```

```python
Summary.groupby("STATE")["OFFENCES AGAINST PROPERTY"].max().plot(kind = "barh", figsize=(10,10))
plt.grid(which="both")
plt.title("OFFENCES AGAINST PROPERTY",fontweight="bold")
plt.xlabel("OFFENCES AGAINST PROPERTY")
plt.ylabel("STATE")
```

```python
Summary.groupby("STATE")[["OFFENCES AGAINST PERSONS","OFFENCES AGAINST PROPERTY","OFFENCES AGAINST LAWFUL AUTHORITY"]].sum().plot(kind = "barh", figsize=(10,10))
plt.grid(which="both")
plt.suptitle("CRIME RATE IN NIGERIA IN 2017",fontweight="bold")
plt.xlabel("RATE OF CRIME")
plt.ylabel("STATE")
```
